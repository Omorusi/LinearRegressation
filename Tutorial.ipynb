```python
import pandas as pd
import numpy as np
import random
from datetime import timedelta, datetime

# Set the number of rows (games) to generate
num_rows = 3000

# Player data setup
players = ['Player_' + str(i) for i in range(1, 3000)]  # 50 players
teams = ['Team_A', 'Team_B', 'Team_C', 'Team_D', 'Team_E']  # Opposing teams

# Define a start date for the games
start_date = datetime(2022, 1, 1)

# Generate random data
data = {
    'Player_ID': [random.choice(players) for _ in range(num_rows)],
    'Game_ID': ['Game_' + str(i) for i in range(1, num_rows + 1)],
    'Date': [start_date + timedelta(days=random.randint(0, 365)) for _ in range(num_rows)],
    'Opponent_Team': [random.choice(teams) for _ in range(num_rows)],
    'Home_Away': [random.choice(['Home', 'Away']) for _ in range(num_rows)],
    'Minutes_Played': np.random.randint(20, 45, size=num_rows),  # Players played between 20 and 45 minutes
    'Points_Scored': np.random.randint(5, 50, size=num_rows),  # Points scored between 5 and 50
    'Assists': np.random.randint(0, 10, size=num_rows),
    'Rebounds': np.random.randint(0, 15, size=num_rows),
    'Turnovers': np.random.randint(0, 6, size=num_rows),
    'Field_Goals_Made': np.random.randint(1, 20, size=num_rows),
    'Field_Goals_Attempted': np.random.randint(5, 30, size=num_rows),
    '3_Pointers_Made': np.random.randint(0, 10, size=num_rows),
    '3_Pointers_Attempted': np.random.randint(0, 15, size=num_rows),
    'Free_Throws_Made': np.random.randint(0, 10, size=num_rows),
    'Free_Throws_Attempted': np.random.randint(1, 15, size=num_rows),
    'Player_Rating': np.random.uniform(0, 10, size=num_rows)  # Random player performance rating between 0 and 10
}

# Create a DataFrame
df = pd.DataFrame(data)

# Add efficiency rating or calculated fields (optional)
df['Shooting_Percentage'] = df['Field_Goals_Made'] / df['Field_Goals_Attempted']
df['Three_Point_Percentage'] = df['3_Pointers_Made'] / df['3_Pointers_Attempted']
df['Free_Throw_Percentage'] = df['Free_Throws_Made'] / df['Free_Throws_Attempted']

# Fill in potential NaN values with 0s (e.g., where attempted = 0)
df.fillna(0, inplace=True)

# Preview the first few rows
print(df.head())

# Save to CSV if needed
# df.to_csv('player_performance_data.csv', index=False)
```
